### About Toronto, CA

Toronto(1) is the largest city in Canada with 2.6 million inhabitants and the capital of the province of Ontario. It is located in the Golden Horseshoe, a region of over 8.1 million inhabitants that extends semicircularly around the western end of Lake Ontario to Niagara Falls. Approximately one-third of the population increase in the whole country lived in this area in the last years. The population of the Census Metropolitan Area increased from 4.1 million in 1992 to 5.6 million in 2011. The Greater Toronto Area had over 6.2 million inhabitants in 2010.

The city is located on the northwestern shore of Lake Ontario, the smallest of the five Great Lakes with an area of 18,960 kmÂ². By incorporating several suburbs that had already merged with Toronto (Etobicoke, Scarborough, York, East York and North York), Toronto was enlarged several times in the late 1990s. The centre with the shopping and banking district is located near the lake. The main shopping street is Yonge Street. Toronto has been Canada's economic centre, and one of the world's leading financial centres since about the 1970s after Montreal had assumed this role for decades.

Toronto is Canada's most important commercial and financial centre and one of the most important in the world. Many banks and investment firms are concentrated in the city's business district on Bay Street. The Toronto Stock Exchange is the eighth largest stock exchange in the world in terms of market capitalization and ranks third in North America. The five largest banks in the country have their headquarters here. In addition, over 40 foreign banks have branches in the city.
